```python
# Define a function that categorizes an hour value into a time-of-day label

def time_of_day(hour):
    if pd.isna(hour):
        return 'Unknown'
    elif 0 <= hour <= 5:
        return 'Night'
    elif 6 <= hour <= 11:
        return 'Morning'
    elif 12 <= hour <= 17:
        return 'Afternoon'
    else:
        return 'Evening'
```

```python
# Apply the time_of_day function to assign a time category to each crime hour

df_final['time_of_day'] = df_final['crime_hour'].apply(time_of_day)
```

```python
# Create a flag indicating whether a weapon was involved based on the weapon description column

df_final['is_weapon_related'] = np.where(df_final['weapon_description'].notna(), 'Yes', 'No')
```

```python
# Define a function that groups crime descriptions into broader crime categories

def crime_category_group(description):
    if pd.isna(description):
        return 'Unknown'
        
    desc = description.lower()
    
    if 'theft' in desc or 'burglary' in desc or 'stolen' in desc:
        return 'Theft / Burglary'
    elif 'assault' in desc or 'robbery' in desc or 'homicide' in desc:
        return 'Violent Crime'
    elif 'vehicle' in desc:
        return 'Vehicle Crime'
    elif 'fraud' in desc or 'forgery' in desc:
        return 'Fraud'
    else:
        return 'Other'
```

```python
# Apply the crime_category_group function to classify each crime into a broad category

df_final['crime_category_group'] = df_final['crime_description'].apply(crime_category_group)
```

```python
# Create a boolean column indicating whether both latitude and longitude values are present

df_final['has_coordinates'] = df_final['latitude'].notna() & df_final['longitude'].notna()
```
